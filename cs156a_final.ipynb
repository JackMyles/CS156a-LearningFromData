The polynomial transform of order $Q = 10$ applied to $\chi$ of dimension $d=2$ results in a $\mathbb Z$ space of dimensionality $\tilde d = 65$ (we don't count the constant coordinates $x_0 = 1$ and $z_0 = 1$). The polynomial transform of order $Q$ applied to some $x \in \chi = (x_1, x_2)$ consists of every variation of $x_1^ix_2^j$ with non-negative integers $i$ and $j$ such that $i+j \leq Q$.

Let us look at the example in which the polynomial transform of order $Q = 4$ is applied to $\chi$ of dimension $d=2$:

$z = (1, x_1, x_2, x_1^2, x_1 x_2, x^2_2, x_1^3, x^2_1 x_2, x_1 x_2^2, x_2^3, x_1^4, x_1^3 x_2, x_1^2 x_2^2, x_1 x_2^3, x_2^4)$

In this exmaple, the dimensionality, $\tilde d$, of the $\mathbb Z$-space is 14. (not counting the constant coordinate $z_0$.) Each monomial corresponds to an entry in the $z$ vector, so the total number of monomials gives us the length of the vector $z$ which is equal to the dimensionality of the $\mathbb Z$ space. In total we have $14$ monomials, so the dimensionality should be $14$, which checks out with the number of components of $z$ (again not counting the constant). In general, if the highest order is $Q$, we have $1 + 2 + ... + (Q+1) = \frac{(Q+1)(Q+2)}{2}$ monomials, so the dimensionality of $\mathbb Z$ is $\frac{(Q+1)(Q+2)}{2} - 1$. In our example we have $Q = 4$ which yields the dimensionality $\frac{(4+1)(4+2)}{2} - 1 = 14$. Therefore the dimensionality of $\mathbb Z$ with $Q = 10$ and dimension $d=2$ is $\frac{(10+1)(10+2)}{2} - 1 = 65$, which is not one of the available choices.

```python
import numpy as np

train = np.loadtxt('features.train')
test = np.loadtxt('features.test')

x_train = train[:, 1:]
digits_train = train[:, 0]
N_train = x_train.shape[0]

x_test = test[:, 1:]
digits_test = test[:, 0]
N_test = x_test.shape[0]

LAMBDA = 1

def get_y(x, digits):
    y = []
    for digit in digits:
        if digit == x:
            y.append(1)
        else:
            y.append(-1)
    return np.array(y)

def linear_regression_reg(Z, y, lambda_val):
    num_columns_Z = Z.shape[1]
    Z_dagger_reg = np.dot(np.linalg.inv(np.dot(Z.T, Z) + lambda_val * np.identity(num_columns_Z)), Z.T)
    w_reg = np.dot(Z_dagger_reg, y)
    return w_reg

def get_Z(X):
    Z = []
    for x1, x2 in X:
        Z.append([1, x1, x2])
    return np.array(Z)

for digit in range(5, 10):
    Z_train = get_Z(x_train)
    Z_test = get_Z(x_test)

    y_train = get_y(digit, digits_train)
    y_test = get_y(digit, digits_test)

    w = linear_regression_reg(Z_train, y_train, LAMBDA)

    E_IN = sum(y_train != np.sign(np.dot(Z_train, w))) / N_train
    E_OUT = sum(y_test != np.sign(np.dot(Z_test, w))) / N_test

    print(f"{digit} versus all:")
    print(f"E_IN is", "%.5f" % E_IN)
    print(f"E_OUT is", "%.5f" % E_OUT)
    print("")
```

```python
from sklearn import svm

def get_y(x):
    y = []
    for p in x:
        x1, x2 = p
        y.append(np.sign(x2 - x1 + 0.25 * np.sin(np.pi * x1)))
    return y

def generate_data(N):
    x = np.random.uniform(-1, 1, (N, 2))
    y = get_y(x)
    return x, y

def SVM(x, y, gamma):
    clf = svm.SVC(kernel='rbf', C=1000000000, gamma=gamma)
    clf.fit(x, y)
    return clf.predict(x)

N = 100
RUNS = 1000
GAMMA = 1.5

E_in_nonzero = 0
for i in range(RUNS):
    x, y = generate_data(N)
    y_predict = SVM(x, y, GAMMA)
    E_in = sum(y_predict != y) / N
    if E_in != 0:
        E_in_nonzero += 1

print(f"You get a dataset that is not separable by the RBF kernel {E_in_nonzero / RUNS * 100}% of the time")
```

```python
def SVM(x, y, gamma):
    clf = svm.SVC(kernel='rbf', C=1000000000, gamma=gamma)
    clf.fit(x, y)
    return clf

def rbf_f(gamma, x, mu):
    return np.exp(-gamma * ((x[0] - mu[0]) ** 2 + (x[1] - mu[1]) ** 2))

def get_phi(x, mus, gamma):
    phi = []
    for p in x:
        row = [1]
        for mu in mus:
            row.append(rbf_f(gamma, p, mu))
        phi.append(row)
    return phi

def get_weight(phi, y):
    return np.dot(np.linalg.pinv(phi), y)

def set_pos(x, mu):
    return np.argmin(np.sqrt(np.sum((x - mu) ** 2, axis=1)))

def lloyd(x, k):
    centers = []
    centers = np.random.uniform(-1, 1, (k, 2))
    size = x.shape[0]
    prev_cluster = np.ones(size)
    curr_cluster = np.zeros(size)
    while not np.all(prev_cluster == curr_cluster):
        prev_cluster = curr_cluster
        curr_cluster = np.apply_along_axis(set_pos, 1, x, centers)
        for c in np.unique(curr_cluster).tolist():
            centers[c] = np.mean(x[curr_cluster == c], axis=0)
    if len(np.unique(curr_cluster)) != k:
        (x, y) = generate_data(size)
        lloyd(x, k)
    return centers

K = 9
RUNS = 100
GAMMA = 1.5
N_train = 100
N_test = 1000

ker_beats_reg = 0
for i in range(RUNS):
    x_train, y_train = generate_data(N_train)
    x_test, y_test = generate_data(N_test)

    # E_out for regular RBF
    centers = lloyd(x_train, K)
    phi_train = get_phi(x_train, centers, GAMMA)
    w = get_weight(phi_train, y_train)
    phi_test = get_phi(x_test, centers, GAMMA)
    y_predict = np.dot(phi_test, w)
    E_out_reg = np.sum(y_predict * y_test < 0) / N_test

    # E_out for kernel RBF
    clf = SVM(x_train, y_train, GAMMA)
    E_out_ker = np.sum(clf.predict(x_test) * y_test < 0) / N_test
    E_in_ker = np.sum(clf.predict(x_train) * y_train < 0) / N_train
    if E_in_ker > 0:
        continue

    if E_out_ker < E_out_reg:
        ker_beats_reg += 1

print(f"Kernel form beats regular form {ker_beats_reg / RUNS * 100}% of the time.")
```

$>$ 75% of the time

```python
def reg_rbf_errors(runs, k, gamma):
    E_in_reg = []
    E_out_reg = []

    for i in range(runs):
        x_train, y_train = generate_data(N_train)
        x_test, y_test = generate_data(N_test)

        centers = lloyd(x_train, k)
        phi_train = get_phi(x_train, centers, gamma)
        w = get_weight(phi_train, y_train)
        y_predict = np.dot(phi_train, w)
        E_in_reg.append(np.sum(y_predict * y_train < 0) / N_train)
        phi_test = get_phi(x_test, centers, gamma)
        y_predict = np.dot(phi_test, w)
        E_out_reg.append(np.sum(y_predict * y_test < 0) / N_test)

    return np.array(E_in_reg), np.array(E_out_reg)

E_in_k9, E_out_k9 = reg_rbf_errors(RUNS, 9, 1.5)
E_in_k12, E_out_k12 = reg_rbf_errors(RUNS, 12, 1.5)

E_in_down = np.sum(E_in_k9 > E_in_k12)
E_out_down = np.sum(E_out_k9 > E_out_k12)

print(f"# of runs where E_in goes down from K = 9 to K = 12 is {E_in_down} and up is {RUNS - E_in_down}")
print(f"# of runs where E_out goes down from K = 9 to K = 12 is {E_out_down} and up is {RUNS - E_out_down}")
```

Since we are assuming that $P(h=f)$ is uniform over $[0,1]$. If we are given no data, the probability that $h=f$ is uniform. Given a single data point $x = 1$ (the chosen person had a heart attack), we know that $f!= 0$, but it is not neccesarily true that $f = 1$. Thus, $P(f = 0 | D) = 0$ and $P(0 < f < 1 | D) != 0$, eliminating answer choices [a] and [d].
To check whether or not $P(h = f | D)$ increases linearly, we can use the formula $$P(h = f | D) = \frac{P(D|h = f)P(h = f)}{P(D)}$$

$P(h=f)$ is a uniform distribution and $P(D | h=f)$ is a linear distribution increasing over [0,1]. This is because, given that one person had a heart attack, the probability that someone with a heart attack is chosen is equal to the probability that any person has a heart attack, which is $h$ and $h = f$. Therefore, as $h$ increases, the posterior, $P(D | h=f)$, increases linearly as well.
