#### 1. The modified Hoeffding Inequality provides a way to characterize the generalization error with a probabilistic bound

#### $P[|E_{in}(g) − E_{out}(g)|$ $>$$ε]$ $≤$ $2Me^{(−2Nε^2)}$

#### for any $ε > 0$. If we set $ε = 0.05$ and want the probability bound $2Me^{−2Nε^2}$ to be at most $0.03$, what is the least number of examples $N$ (among the given choices) needed for the case $M = 1$?

[a] 500

[b] 1000

[c] 1500

[d] 2000

[e] More examples are needed.

#### Answer:
B

$ε = 0.05$

$M = 1$

$0.03 = 2e^{-2N(0.05)^2}$

$0.015 = e^{-0.005N}$

$ln(0.015) = ln(e^{-0.005N})$

$-4.1997 = -0.005N$

$N = 839.941 \approx 1000$

#### 7. Which of the following is the growth function $m_H(N)$ for the “2-intervals” hypothesis set?

[a] $N+1 \choose 4$

[b] ${N+1 \choose 2} + 1$

[c] ${N+1 \choose 4}$ $+ {N+1 \choose 2} + 1$

[d] ${N+1 \choose 4} + {N+1 \choose 3} + {N+1 \choose 2} + {N+1 \choose 1} + 1$

[e] None of the above

#### Answer:
C

We know that there are $N+1$ spots we could place the interval ends (slide 12 of lecture 5). For a single interval, we must choose 2 of these spots $N+1 \choose 2$. This covers any dichotomy where the points within the intervals are all adjacent to each other, so only one interval is needed. For the dichotomies where both intervals are needed, we must choose 4 of the $N+1$ spots $N+1 \choose 4$. There is also one dichotomy where the interval ends are in the same spot so they don't contain any points. Thus, the total growth function would be $m_H(N) = {N+1 \choose 4}$ $+ {N+1 \choose 2} + 1$.

#### 10. Compute the growth function $m_H(N)$ for the learning model made up of two concentric circles around the origin in $\mathbb{R}^2$. Specifically, $H$ contains the functions which are $+1$ for $a^2 ≤ x_1^2 + x_2^2 ≤ b^2$ and $−1$ otherwise, where $a$ and $b$ are the model parameters. The growth function is:

[a] $N+1$

[b] ${N+1 \choose 2} + 1$

[c] ${N+1 \choose 3} + 1$

[d] $2N^2 + 1$

[e] None of the above

#### Answer:
B

The only thing that matters for determining whether or not the point is within the area of the concentric circles is the distance of the point from the origin. Therefore the problem is essentially the same as the problem involving a single interval containing points along a line. The ring is the same as the interval on the line. The growth function for the single interval hypothesis set is ${N+1 \choose 2} + 1$, so the growth function for this hypothesis set is also ${N+1 \choose 2} + 1$.
